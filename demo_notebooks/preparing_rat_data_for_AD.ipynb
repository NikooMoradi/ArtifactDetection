# Preparing Rat Data
### This notebook saves the rat recordings in a clean format, suitable for using ArtifactDetection.
### Make sure recording files are in .npy and brain state files are in .pkl before applying artifactdetection functions


Apply the following functions to convert recording files from .dat files to .npy, and brainstate files from .xls to .pkl

## Importing Packages

```python
import os 
import sys
import numpy as np
import pandas as pd
sys.path.insert(0, "/home/s2864332/MySYNGAP/ArtifactDetection")
import artifactdetection as ad
```

## Defining Paths

```python
# Recordings paths
raw_rec_folder = "/home/s2864332/SYNGAP_Rat_Data/raw/raw_recordings_all"
output_rec_folder = ".../formatted_rec" # Adjust the path if needed
if not os.path.exists(output_rec_folder):
    os.makedirs(output_rec_folder, exist_ok=True)

# Brain states paths
brain_state_folder = "/home/s2864332/SYNGAP_Rat_Data/raw/raw_brain_states"
output_brain_state_folder = ".../formatted_br" # Adjust the path if needed
if not os.path.exists(output_brain_state_folder):
    os.makedirs(output_brain_state_folder)

# Excel metadata path anf animal IDs
xlsx_path = "/home/s2864332/SYNGAP_Rat_Data/Recordings Samp Start-End TAINI - ALL DAYS.xlsx"
animal_IDs = [file for file in os.listdir(raw_rec_folder)]
```

## Preparing the metadata excell file

```python
metadata = pd.read_excel(xlsx_path, header=None)
metadata = metadata.iloc[6:48, 2:8] # Adjusted to match the actual data range
metadata.columns = ['Animal ID', 'GENO', 'FILE NAME','DAY', 'Start Sample', 'End Sample'] # Assign column names
metadata["Animal ID"] = metadata["Animal ID"].ffill() # Fill down Animal ID
metadata["GENO"] = metadata["GENO"].ffill() # Fill down genotype

metadata.reset_index(drop=True, inplace=True) 
metadata = metadata.drop("DAY", axis=1) # Drop DAY column
```

## 1. Convert .dat to .npy using the Preprocess class

### Looping through each animal, load the .dat recording, convert it into .npy format, save it in output directory

```python
for animal_id, group in metadata.groupby("Animal ID"):
    

    file_name = group["FILE NAME"].iloc[0]   # same for BL1 and BL2
    if pd.isna(file_name):
        file_name = group["FILE NAME"].iloc[1]
        
    raw_folder = os.path.join(raw_rec_folder, animal_id)

    output_name = f"{animal_id}_BASELINE.npy"
    output_path = os.path.join(raw_rec_folder, output_name)

    print(30*"=")
    print(f"Processing: {animal_id}")
    save_as = animal_id + "_BASELINE"
    preprocess = ad.Preprocess(raw_folder = raw_folder, raw_file = file_name, save_fold_path = output_rec_folder, save_as = save_as)
    preprocess.convert_dat_to_npy()
```

## 2. Convert .xls to .pkl 

### Looping through each animal, load the brain states, convert it into .pkl format, save it in output directory

```python
unformatted_br_files = [file for file in os.listdir(brain_state_folder)] #list of file names to feed to preprocess

for file in unformatted_br_files:
    print(30*"=")
    br_id = file[0:9]
    print(f"Processing: {br_id}")
    print(file)
    if file.endswith(".xls"):
        ad.reformat_br_file(unformatted_folder = brain_state_folder, unformatted_file = file, save_folder = output_brain_state_folder,
                       save_as = br_id)
```
